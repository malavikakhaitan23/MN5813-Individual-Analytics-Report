### 6.3 Load data

Here, the raw data files are loaded into the Python environment. Loading data is vital since it turns external files into in-memory objects for inspection, cleaning, and analysis. 

```python
covid = pd.read_csv("Covid 19 Dataset.csv")
```

```python
gdp = pd.read_csv(
    "GDP per capita Dataset.csv",
    skiprows=4
)
```

```python
unemployment = pd.read_csv(
    "Unemployment Rate Dataset.csv",
    skiprows=4
)
```

```python
health = pd.read_csv(
    "Healthcare Expenditure Dataset.csv",
    skiprows=4
)
```

### 6.4 Convert into DataFrame

After the data is loaded, it is transformed to Pandas DataFrames. DataFrames have a tabular format that is useful for filtering, combining, manipulating, and analysing data. This part formalises the shift from raw input to a structured analytical format.

```python
# Keep only the columns needed for analysis and make a clean copy
covid = covid[[
    "country",
    "date",
    "population",
    "total_cases",
    "total_deaths",
    "total_vaccinations",
    "total_tests"
]].copy()

# Convert the date column to datetime
covid["date"] = pd.to_datetime(covid["date"], errors="coerce")

# Remove any rows where the date could not be parsed
covid = covid.dropna(subset=["date"])

# Extract the calendar year for yearly aggregation
covid["year"] = covid["date"].dt.year

# Compute cases and deaths per 100,000 people
covid["cases_per_100k"] = (covid["total_cases"] / covid["population"]) * 100000
covid["deaths_per_100k"] = (covid["total_deaths"] / covid["population"]) * 100000

# Aggregate data to yearly level for each country
covid_yearly = (
    covid
    .groupby(["country", "year"], as_index=False)
    .agg({
        "population": "max",
        "total_cases": "max",
        "total_deaths": "max",
        "total_vaccinations": "max",
        "total_tests": "max",
        "cases_per_100k": "max",
        "deaths_per_100k": "max"
    })
)

# Create a weighted COVID-19 burden index combining cases and deaths
covid_yearly["covid_burden_index"] = (
    covid_yearly["cases_per_100k"] * 0.6 +
    covid_yearly["deaths_per_100k"] * 0.4
)

# Scale the burden index for easier interpretation or visualisation
covid_yearly["burden_scaled"] = covid_yearly["covid_burden_index"] / 100
```

```python
# Reshape GDP data from wide format (years as columns) to long format
gdp_long = gdp.melt(
    id_vars=["Country Name", "Country Code"],
    var_name="year",
    value_name="gdp_per_capita"
)

# Reshape unemployment data into long format 
unemployment_long = unemployment.melt(
    id_vars=["Country Name", "Country Code"],
    var_name="year",
    value_name="unemployment_rate"
)

# Reshape health expenditure data into long format
health_long = health.melt(
    id_vars=["Country Name", "Country Code"],
    var_name="year",
    value_name="health_expenditure_pct_gdp"
)

# Convert the 'year' column from string to numeric for all three reshaped datasets
for df in [gdp_long, unemployment_long, health_long]:
    df["year"] = pd.to_numeric(df["year"], errors="coerce")
```

```python
# Merge GDP and unemployment data 
world_bank = (
    gdp_long
    .merge(
        unemployment_long,
        on=["Country Name", "Country Code", "year"],
        how="left"
    )

# Merge health expenditure data into the combined dataset
    .merge(
        health_long,
        on=["Country Name", "Country Code", "year"],
        how="left"
    )
)

# Rename "Country Name" to simply "country" for consistency with covid_yearly
world_bank = world_bank.rename(columns={
    "Country Name": "country"
})

# Merge the COVID yearly dataset with the World Bank indicators
final_df = covid_yearly.merge(
    world_bank,
    on=["country", "year"],
    how="left"
)
```

#### 7.1.7 Country Identifier Harmonisation

Merging datasets from several sources necessitates precise alignment of country identities. Countries' names are checked for anomalies such as alternate spellings, abbreviations, or geopolitical disparities. Where required, manual adjustments are made to guarantee proper alignment.

This step is crucial for avoiding silent data loss during merging and is carefully documented to ensure repeatability. While time-consuming, identifier harmonisation greatly enhances data integrity.

```python
# Extract the list of unique country names from the COVID dataset
covid_countries = (
    covid["country"]
    .dropna()
    .unique()
)

# Extract the list of unique country names from the World Bank dataset
wb_countries = (
    world_bank["country"]
    .dropna()
    .unique()
)

# Convert both lists to sets for easy comparison
covid_set = set(covid_countries)
wb_set = set(wb_countries)

# Identify countries that appear only in the COVID dataset
covid_only = covid_set - wb_set

# Identify countries that appear only in the World Bank dataset
wb_only = wb_set - covid_set

# Print mismatched country names from both datasets
print("Only in COVID dataset:", sorted(covid_only))
print("Only in World Bank dataset:", sorted(wb_only))

# Define a mapping to standardise country names 
country_mapping = {
    "United States": "United States of America",
    "South Korea": "Korea, Rep.",
    "Russia": "Russian Federation",
    "Iran": "Iran, Islamic Rep.",
    "Vietnam": "Viet Nam",
    "Republic of Korea": "South Korea",
    "Korea, South": "South Korea",
    "Korea, Rep.": "South Korea"
}

# Apply the country name corrections to the COVID dataset
covid["country"] = covid["country"].replace(country_mapping)

# Recompute the updated set of COVID country names after renaming
covid_set_updated = set(covid["country"].dropna().unique())

# Identify any remaining unmatched countries after applying the mapping
remaining_unmatched = covid_set_updated - wb_set

print("Remaining unmatched countries:", sorted(remaining_unmatched))

# Identify countries that exist in both datasets after harmonisation
unified_countries = sorted(
    covid_set_updated.intersection(wb_set)
)

# Convert the unified list of countries into a dataframe
unified_country_df = pd.DataFrame(
    unified_countries,
    columns=["country"]
)

# Preview the unified country list
unified_country_df.head(190)
```

#### 7.1.8 Outlier Detection and Treatment

Outliers are identified using both eye inspection and descriptive statistics. Instead of automatically rejecting extreme numbers, the analysis determines if outliers are actual observations or possible reporting mistakes.

Outliers are often kept because extreme outcomes are substantively significant in the context of a worldwide pandemic. However, their impact on interpretation is fully addressed in the analysis section.

```python
# Generate descriptive statistics for cases and deaths per 100k
final_df[["cases_per_100k", "deaths_per_100k"]].describe()

def flag_outliers(series):
    """
    Identify outliers in a numeric pandas Series using the IQR method.

    Parameters
    ----------
    series : pandas.Series
        The numeric column for which outliers should be detected.

    Returns
    -------
    pandas.Series (bool)
        A boolean mask where True indicates an outlier and False indicates
        a non-outlier, based on the 1.5 * IQR rule.
    """
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (series < lower_bound) | (series > upper_bound)

# Apply the outlier flagging function to cases and deaths per 100k
final_df["cases_outlier"] = flag_outliers(final_df["cases_per_100k"])
final_df["deaths_outlier"] = flag_outliers(final_df["deaths_per_100k"])

# Display the top 10 countries with the highest case-rate outliers
final_df.loc[final_df["cases_outlier"], 
              ["country", "cases_per_100k"]].sort_values("cases_per_100k", ascending=False).head(10)
```

### 14.2 Appendix B: Supplementary Methodological Notes and Code Structure

This appendix provides additional methodological detail supporting the analytical decisions made in the project. These notes clarify assumptions, justify analytical choices, and explain the reasoning behind key steps in the workflow.

#### B1. Rationale for Population Normalisation

COVID19 case and death counts vary dramatically across countries due to differences in population size. To ensure meaningful crosscountry comparison, all epidemiological indicators were converted into rates per 100,000 and per 10,000 people.  

This approach:

- reduces distortion caused by largepopulation countries  
- allows fair comparison between small and large nations  
- aligns with publichealth reporting standards  

Normalisation does not eliminate structural biases (e.g., age distribution, co-morbidities), but it significantly improves analytical comparability.

#### B2. Construction of the COVID19 Burden Index

The COVID19 burden index was created to summarise pandemic severity using a transparent, reproducible formula:

- 60% weight on cases per 100k
- 40% weight on deaths per 100k

This weighting reflects:

- the higher frequency and variability of case data  
- the greater severity but lower volume of death data  

The index is not intended as a clinical severity measure. Its purpose is to provide a comparative analytical metric that captures both infection intensity and mortality burden.

#### B3. Yearly Aggregation of COVID19 Data

COVID19 data from OWID is recorded daily, whereas World Bank indicators are annual. To align the datasets:

- daily COVID19 data was aggregated to yearly maxima
- this preserves cumulative totals (cases, deaths, vaccinations)  
- it ensures temporal compatibility with economic indicators  

This method avoids mixing incompatible time scales and maintains consistency across all variables.

#### B4. Handling Missing Data

The project uses listwise deletion for countries with insufficient data coverage. This decision was made because:

- imputation introduces assumptions that may distort crosscountry comparisons  
- missingness is often structural (e.g., limited testing capacity)  
- transparency is prioritised over completeness  

This limitation is acknowledged in the main report and is consistent with exploratory analytics practice.

#### B5. Reshaping and Merging Logic

World Bank datasets are provided in wide format (years as columns). To integrate them with the COVID19 dataset:

- each dataset was reshaped into long format  
- merging was performed on countryyear 

This process creates a unified analytical dataset with aligned temporal and geographic identifiers.

#### B6. Justification for Visual Analytics Approach

Visual analytics was chosen as the primary method because the project is exploratory rather than predictive. 

Visualisation supports:

- pattern recognition  
- distributional analysis  
- identification of outliers  
- comparison across countries and indicators  

Techniques such as small multiples, scatterplots, and histograms allow multi-dimensional relationships to be examined without imposing restrictive model assumptions. This approach aligns with Tukeys principles of exploratory data analysis.

#### B7. Risk Classification Method

Countries were classified into *Low*, *Medium*, and *High* risk groups using a mediansplit approach applied to the burden index.  

This method:

- avoids arbitrary thresholds  
- ensures balanced group sizes  
- supports comparative vulnerability analysis  

Only the resulting risk_level variable is used in the analysis; no separate threshold variable is stored.

#### B8. Limitations of CrossCountry Comparison

Crosscountry COVID19 analysis is inherently constrained by:

- inconsistent reporting standards  
- differences in testing capacity  
- demographic variation  
- political and institutional differences  
- variation in healthcare infrastructure  

These limitations do not invalidate the analysis but require cautious interpretation of observed patterns.
