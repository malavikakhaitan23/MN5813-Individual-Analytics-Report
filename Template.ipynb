### 6.3 Load data

```python
covid = pd.read_csv("Covid 19 Dataset.csv")
```

```python
gdp = pd.read_csv(
    "GDP per capita Dataset.csv",
    skiprows=4
)
```

```python
unemployment = pd.read_csv(
    "Unemployment Rate Dataset.csv",
    skiprows=4
)
```

```python
health = pd.read_csv(
    "Healthcare Expenditure Dataset.csv",
    skiprows=4
)
```

### 6.4 Convert into DataFrame

```python
covid = covid[[
    "country",
    "date",
    "population",
    "total_cases",
    "total_deaths",
    "total_vaccinations",
    "total_tests"
]].copy()

covid["date"] = pd.to_datetime(covid["date"], errors="coerce")

covid = covid.dropna(subset=["date"])

covid["year"] = covid["date"].dt.year

covid_yearly = (
    covid
    .groupby(["country", "year"], as_index=False)
    .max()
)
```

```python
gdp_long = gdp.melt(
    id_vars=["Country Name", "Country Code"],
    var_name="year",
    value_name="gdp_per_capita"
)

unemployment_long = unemployment.melt(
    id_vars=["Country Name", "Country Code"],
    var_name="year",
    value_name="unemployment_rate"
)

health_long = health.melt(
    id_vars=["Country Name", "Country Code"],
    var_name="year",
    value_name="health_expenditure_pct_gdp"
)

for df in [gdp_long, unemployment_long, health_long]:
    df["year"] = pd.to_numeric(df["year"], errors="coerce")
```

```python
world_bank = (
    gdp_long
    .merge(
        unemployment_long,
        on=["Country Name", "Country Code", "year"],
        how="left"
    )
    .merge(
        health_long,
        on=["Country Name", "Country Code", "year"],
        how="left"
    )
)

world_bank = world_bank.rename(columns={
    "Country Name": "country"
})

final_df = covid_yearly.merge(
    world_bank,
    on=["country", "year"],
    how="left"
)
```

```python
final_df[["cases_per_100k", "deaths_per_100k"]].describe()

def flag_outliers(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return (series < lower_bound) | (series > upper_bound)

final_df["cases_outlier"] = flag_outliers(final_df["cases_per_100k"])
final_df["deaths_outlier"] = flag_outliers(final_df["deaths_per_100k"])

final_df.loc[final_df["cases_outlier"], 
              ["country", "cases_per_100k"]].sort_values("cases_per_100k", ascending=False).head(10)
```

### 10.1 Discussion of Findings in Context

The outcomes of this investigation demonstrate the complexities of evaluating global crises using secondary data. While the COVID-19 pandemic is frequently described in terms of headline numbers like case counts and death tolls, the study shows that these measures are heavily influenced by economic capacity, healthcare infrastructure, and data reporting standards. As a result, apparent cross-country disparities typically reflect structural rather than solely epidemiological factors.

This highlights the significance of contextualisation in business analytics. Analytics is more than just processing massive numbers; it's also about evaluating them within organisational, institutional, and socioeconomic frameworks. The experiment demonstrates how combining health and economic data enables analysts to go beyond superficial comparisons and gain more significant insights.

### 10.2 Implications for Business Analytics Practice

One of the project's main implications is the use of multidimensional analytics in tackling complicated, real-world challenges. Single-variable dashboards or league tables may be appealing for their simplicity, but they risk oversimplifying phenomena caused by interacting elements. In contrast, integrating epidemiological data with economic and healthcare factors allows for richer analytical narratives.

For business analysts, the project highlights the practical utility of Pandas-based data wrangling, exploratory visualisation, and comparative analytical methods. These techniques and approaches have a wide range of applications outside of public health, including supply chain resilience, financial risk assessment, and organisational performance analysis during disruptions.

### 10.3 Methodological Limitations

Despite its virtues, the project has numerous significant drawbacks. First, the study is solely based on secondary data sources. While both Our World in Data and the World Bank are credible archives, they collect information from national reporting bodies that have different criteria, definitions, and motivations. This reduces the comparability of certain indicators between countries.

Second, the analysis is mostly descriptive and exploratory. There are no causal assertions made about the effect of economic factors on COVID-19 outcomes. Establishing causality would need more advanced approaches, such as panel regression models, instrumental variables, or quasi-experimental designs, which are outside the scope of this assignment.

Third, temporal alignment is a hurdle. World Bank economic statistics are frequently released annually and with lag, but COVID-19 data is available on a daily basis. Despite using adequate alignment procedures, this mismatch limits the precision with which dynamic interactions may be analysed.

### 10.4 Data Quality and Bias Considerations

Data quality is a major consideration throughout the investigation. Missing data in testing and healthcare spending variables is consistently associated with lower-income nations, potentially biassing global summaries and comparisons. While data cleaning processes attempted to address these difficulties, they could not completely eradicate structural inequities in data availability.

Political and institutional considerations may also impact reporting standards, notably for case and death statistics. These variables are difficult to measure and hence stay implicit in the study. Recognising these biases is critical for good analytics practice.

### 10.5 Opportunities for Future Work

The project's findings point to various options for further study and analytical development. First, the incorporation of demographic characteristics such as age structure or urbanisation rates may improve the understanding of mortality results. Second, more granular policy data, such as lockdown stringency indexes, might be used to evaluate the effectiveness of nonpharmaceutical measures.

Future research might use longitudinal models to better capture temporal dynamics and account for unobserved variability between nations. Machine learning algorithms might also be investigated for pattern discovery, albeit their use would need to be carefully justified given the quality and structure of the underlying data.

Finally, expanding the research to include post-pandemic recovery measures such as economic growth or labour market results will provide light on the pandemic's long-term implications and the role of institutional resilience.

### 10.6 Summary

In summary, this debate highlights both the importance and limitations of exploratory analytics applied to global crisis data. The study illustrates how intentional data integration, clear methodology, and critical interpretation may generate decision-relevant insights while simultaneously acknowledging the limitations imposed by data quality and methodological scope.
